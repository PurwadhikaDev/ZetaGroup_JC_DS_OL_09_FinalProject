##Project Background

Berdasarkan data [Portuguese Banking Association](https://www.apb.pt/publications_and_papers/publications/portuguese_banking_sector_overview/), diketahui bahwa industri perbankan Portugal mengalami peningkatan saldo deposito. Hal ini membuat rasio pinjaman terhadap deposito atau biasa disebut dengan LDR (*Loan to Deposit Ratio*) mengalami penurunan. Dikutip dari Stephen Buschbom dalam [Assessing Bank Risk Using the Loan-to-Deposit Ratio](https://www.trepp.com/trepptalk/assessing-bank-risk-using-loan-deposit-ratio#:~:text=The%20LDR%20is%20often%20used,hand%20to%20cover%20unexpected%20losses), diketahui bahwa LDR sering digunakan sebagai indikator tingkat risiko bank, dimana rasio yang tinggi menunjukkan bahwa bank mengambil lebih banyak risiko karena memiliki cadangan kas yang lebih sedikit untuk menutupi kerugian yang tidak terduga.

Data peforma industri Bank Portugal ini menyoroti kondisi yang menguntungkan dalam sektor industri perbankan karena LDR-nya yang mengalami penurunan, hal ini menunjukkan bahwa bank-bank Portugal telah memperkuat posisi likuiditas-nya sehingga level risikonya menjadi lebih rendah dari sebelumnya dan dapat memainkan peran penting dalam mendukung perekonomian.  Hal ini memungkinkan bank-bank untuk meningkatkan kemampuan mereka dalam menyediakan pinjaman kepada individu, bisnis, dan sektor-sektor lain dalam perekonomian. Ketersediaan likuiditas yang cukup dapat membantu dalam mendorong aktivitas ekonomi yang sehat, seperti investasi, konsumsi, dan pertumbuhan bisnis.

Hal ini menunjukkan bahwa deposito merupakan salah satu sumber dana yang penting bagi bank dalam menjalankan operasionalnya. Oleh karena itu, **kampanye pemasaran** yang menargetkan peningkatan jumlah deposito dapat memberikan manfaat besar bagi bank.

Prediksi ini sangat penting karena berpotensi berdampak pada keseluruhan operasional dan keberhasilan bank dalam **memitigasi risiko** hal yang tak terduga.

## Metric Evaluation

image.png

**FN (False Negative)** : Model salah memprediksi bahwa seorang nasabah tidak akan melakukan deposito, padahal sebenarnya nasabah tersebut akan melakukan deposito <br>
**--> Konsekuensi** : Bank kehilangan peluang untuk menarik nasabah potensial yang sebenarnya akan merespons positif terhadap kampanye.
<br><br>
**FP (False Positive)** : Model salah memprediksi bahwa seorang nasabah akan melakukan deposito, padahal sebenarnya nasabah tidak akan melakukan deposito <br>
**--> Konsekuensi** : Bank mengalokasikan sumber daya tambahan kepada nasabah yang sebenarnya tidak berminat
<br><br>
**TN (True Negative)** : Seorang nasabah tidak akan melakukan deposito dan sebenarnya nasabah tersebut memang tidak akan melakukan deposito <br>
**--> Manfaat** : Bank menghindari pengeluaran yang tidak perlu dan menghemat sumber daya untuk nasabah yang tidak berminat atau tidak akan merespons positif
<br><br>
**FP (True Positive)** : Seorang nasabah akan melakukan deposito dan sebenarnya nasabah tersebut memang akan melakukan deposito.<br>
**--> Manfaat** : Bank berhasil menarik nasabah yang berpotensi merespons positif dan berkontribusi pada pendapatan dan pertumbuhan bisnis.
<br><br>
Berdasarkan tujuan dari project untuk memprediksi sebanyak - banyaknya nasabah mana yang akan melakukan deposito (TP) dan yang tidak (TN), maka metric utama yang akan digunakan adalah accuracy.<br><br>

### Default, Housing, Loan, y

Diketahui pada kolom 'default', 'housing', 'loan', 'y' pada dasarnya memiliki value binary (yes/no). Untuk mempermudah analisis maka value 'yes' dan 'no' akan diubah menjadi "1" (yes) dan "0" (no).

### Default

Berkaitan dengan kolom default yang pada dasarnya berupa binary jika semua value terdefinisi (yes/no), maka 'unknown' value akan di impute menjadi 1 atau 0. Untuk itu perlu dilakukan eksplorasi data lebih lanjut.

```python
def plot_deposit_frequencies(data, column, figsize=(8, 5), variable_order=0):
    # Calculate the counts of 'yes' values for each unique variable
    deposit_counts = data.groupby(column)['y'].value_counts().unstack()

    # Construct the table with unique values and deposit counts for each value
    table = deposit_counts.copy()

    # Sort the table in descending order based on the 'y' = Yes column
    table_sorted = table.sort_values(1, ascending=False)

    # Create a new figure and axes with the desired size
    fig, ax = plt.subplots(figsize=figsize)

    # Plot the stacked bar chart
    table_sorted.plot.bar(stacked=True, ax=ax)

    # Set the title and labels
    ax.set_title(f'Deposit Frequencies by {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')
    ax.legend(loc='upper right')

    # Display the chart
    plt.show()

    # Print the sorted table
    print('----Proportion Table----')
    print(table_sorted)

    # Calculate and print the table with percentages
    table_percentage = table_sorted.divide(table_sorted.sum(axis=1), axis=0) * 100
    print('')
    print('----Percentage Table----')

    if variable_order == 1:
        print(table_percentage[1])

        # Calculate Faktorisasi Percentage
        faktorisasi_percentage = (table_sorted[1] * table_percentage[1]).round(2)
        print('')
        print('----Faktorisasi Percentage----')
        print(faktorisasi_percentage)

    else:
        print(table_percentage)
```

## Splitting

```python
# Define dependent variable as target / label of prediction
y = df_marketing['y']

# Define independent variable as feature
x = df_marketing.drop(columns = ['y'])

# Split Dataset for Training (80%) and Testing (20%)
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=1,
    stratify=y
) # 0.2 is 20% of the data as a tester  |  1 just for random | stratify for sampe proportion y
```

## Encoding

Akan dilakukan encoding dengan beberapa metode untuk feature categorical:
- Onehot, dengan mempertimbangkan sedikitnya unique value.
  - marital
  - contact
  - poutcome
- Binary, dengan mempertimbangkan banyaknya unique value.
  - job
  - education
  - month
  - day_of_week

```python
# One Hot Encoding
one_hot_var = ['marital', 'contact', 'poutcome']
one_hot_encoder = OneHotEncoder()

# Binary Encoding
binary_var = ['job', 'education', 'month', 'day_of_week']
binary_encoder = ce.BinaryEncoder()

# Combine preprocess in one table
preprocess= ColumnTransformer([
                                ('one_hot', one_hot_encoder, one_hot_var),
                                ('binary', binary_encoder, binary_var),
                                ], remainder='passthrough')
```

```python
# Set the option to show all columns
pd.set_option('display.max_columns', None)

# Combine spliting data into preprocessing usin .fit_transform
x_train_prep = preprocess.fit_transform(x_train)
x_test_prep = preprocess.transform(x_test)

x_train_prep = pd.DataFrame(x_train_prep)
x_test_prep = pd.DataFrame(x_test_prep)

x_train_prep
```

```python
# Getting Feature Names from one_hot_encoding
name_one_hot = list(preprocess.transformers_[0][1].get_feature_names_out())

# Getting Feature Names from binary_encoding
name_binary = list(preprocess.transformers_[1][1].get_feature_names_out())

# Adjust sequence in naming based on ColumnTransformer
feature_names = name_one_hot + name_binary + ['age','default','housing','loan','duration','campaign','pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
feature_names
```

## Model

Sebagaimana diketahui bahwa tujuan dari model adalah untuk memprediksi nasabah melakukan deposit atau tidak, maka model yang digunakan adalah klasifikasi. Tentunya hal lain yang perlu dipertimbangkan adalah data bersifat imbalance dan range valuenya sangat beragam, maka data perlu dipertimbangkan untuk dilakukan perlakuan data imbalance dan scale.

Model yang dicoba:
- Logistic Regression <br>
  Memperkirakan kemungkinan sebuah instance milik kelas tertentu menggunakan fungsi logistik. <br>
- K-Nearest Neighbors (KNN) <br>
  Mengukur jarak antara instance baru dan instance tetangganya untuk menentukan kelasnya. <br>
- Decision Tree Classifier <br>
  Membagi dataset berdasarkan nilai fitur untuk membuat subkelompok yang lebih homogen dalam hal variabel target. <br>
- Random Forest Classifier <br>
  Menciptakan "hutan" pohon keputusan, dimana setiap pohon secara independen memprediksi kelas instance, Prediksi akhir dibuat dengan voting mayoritas. <br>
- AdaBoost Classifier <br>
  Secara iteratif melatih pengklasifikasi yang lemah pada subset data yang berbeda, memberi bobot lebih pada instance yang salah klasifikasi di setiap iterasi. Prediksi akhir dibuat dengan menggabungkan prediksi dari semua pengklarifikasi dengan bobot telah diberikan untuk kinerja yang lebih kuat <br>
- Gradient Boosting Classifier <br>
  Membangun model secara iteratif, dimana setiap pohon baru mengkoreksi kesalahan yang dibuat dengan menjumlahkan prediksi dari semua pohon, ditimbang dengan kontribusi masing-masing <br>
- XGBClassifier <br>
  Menyempurnakan algoritma Gradient Boosting dengan fitur tambahan (proses paralel, menghupus cabang atau node yang tidak menggeneralisasi ke data yang tidak terlihat)

Scaler method yang dicoba:
- MinMaxScaler <br>
  Mengubah skala data menjadi rentang tertentu, seperti 0 hingga 1. Mengurangi nilai minimum dari setiap data. Ini membantu memastikan bahwa semua fitur memiliki skala yang serupa, mencegah fitur tunggal mendominasi proses analisis atau pelatihan model, dan memungkinkan perbandingan dan prediksi yang lebih akurat. <br>
- StandardScaler <br>
  Mengubah data sehingga memiliki rata-rata nol dan standar deviasi satu. Dengan cara mengurangi rata-rata dari setiap nilai data dan membaginya dengan standar deviasi. <br>
- RobustScaler <br>
  Mengubah skala data dengan mempertahankan ketahanan terhadap pencilan (outliers). Dengan mengurangi median dari setiap nilai data dan membaginya dengan selisih antara kuartil ketiga (Q3) dan kuartil pertama (Q1). Scaler ini membantu menjaga konsistensi skala data bahkan dalam hadirnya outlier, cocok digunakan saat ada nilai ekstrim dalam data.

Imbalance treatment yang dicoba:
- RandomOverSampler <br>
  Digunakan untuk mengatasi ketidakseimbangan data dengan menambahkan data acak dari kelas minoritas sehingga proporsi antara kelas mayoritas dan minoritas menjadi lebih seimbang.
- SMOTE <br>
  Metode ini menciptakan data sintetis baru dari kelas minoritas dengan cara menggabungkan tetangga terdekat. Ini membantu meningkatkan perwakilan kelas minoritas dalam data yang tidak seimbang, sehingga algoritma pembelajaran mesin dapat mempelajari pola dari kelas minoritas dengan lebih baik
- RandomUnderSampler <br>
  Metode yang mengurangi jumlah sampel dari kelas mayoritas secara acak untuk mengatasi ketidakseimbangan data.

```python
# Define the parameter grid for GridSearchCV
param_grid = [
    {
        'scaler': [None,MinMaxScaler(), StandardScaler(), RobustScaler()],
        'imbalance':[None,
                     RandomOverSampler(random_state=1),
                     SMOTE(random_state=1),
                     RandomUnderSampler(random_state=1)],
        'classifier' : [LogisticRegression(),
                        KNeighborsClassifier(),
                        RandomForestClassifier(),
                        DecisionTreeClassifier(),
                        AdaBoostClassifier(),
                        GradientBoostingClassifier(),
                        XGBClassifier()]
    }
]

# Define Model with Scaler and Imbalance Treatment
pipeline = Pipeline([
    ('scaler', StandardScaler()), # Default Scaler
    ('imbalance', SMOTE(random_state=1)), # Default Imbalance
    ('classifier', LogisticRegression()) # Default Model
])

# Create GridSearchCV instance
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Start the timer
start_time = time.time()

# Fit x_train_prep & y_train
grid_search.fit(x_train_prep, y_train)

# Calculate the elapsed time
elapsed_time = time.time() - start_time

# Convert elapsed time to minutes and seconds
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)

# Get the best estimator and its performance scores
pipeline = grid_search.best_estimator_
best_scores = grid_search.best_score_

# Print the results
print("Best Scaler:", pipeline.named_steps['scaler'])
print("Best Imbalance:", pipeline.named_steps['imbalance'])
print("Best Model:", pipeline.named_steps['classifier'])
print("Best Scores:", best_scores)

# Print the elapsed time
print("Elapsed Time:", minutes, "minutes", seconds, "seconds")
```

### Result with Data Train

Diketahui bahwa model terbaik sementara adalah:

- Tanpa menggunakan perlakuan tambahan untuk data imbalance

Idenya adalah dengan menggabungkan prediksi dari semua model ini, dengan mempertimbangkan kekuatan dan kelemahannya, kita dapat membuat model prediksi yang kuat dan akurat.

Ini seperti memiliki sekelompok siswa dengan kekuatan dan kelemahan berbeda yang bekerja sama untuk menjawab pertanyaan.

image.png

- Parameter:
  - model__learning_rate: 0.1 <br>
    Ini mengontrol dampak dari setiap pohon individu selama proses pembelajaran. Tingkat pembelajaran yang kecil memberikan akurasi yang lebih besar tetapi membutuhkan waktu lebih lama untuk berlatih, sementara tingkat pembelajaran yang besar melatih lebih cepat tetapi dapat menyebabkan overfitting.
  - model__max_depth: 3 <br>
    Ini membatasi kompleksitas masing-masing pohon. Nilai yang lebih tinggi memungkinkan model mempelajari pola yang rumit tetapi dapat menyebabkan overfitting, sedangkan nilai yang lebih rendah membuat model lebih kuat tetapi berpotensi kurang akurat.
  - model__n_estimators: 250 <br>
    Menentukan berapa banyak pohon yang digabungkan dalam model. Lebih banyak pohon umumnya mengarah pada kinerja yang lebih baik, tetapi hal itu disertai dengan biaya waktu pelatihan yang meningkat dan risiko overfitting.



*   MinMaxScaler <br>
    Mengubah skala data menjadi rentang tertentu, seperti 0 hingga 1. Mengurangi nilai minimum dari setiap data. Ini membantu memastikan bahwa semua fitur memiliki skala yang serupa, mencegah fitur tunggal mendominasi proses analisis atau pelatihan model, dan memungkinkan perbandingan dan prediksi yang lebih akurat.
